# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. 
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#open chrome browser and the URL
driver=webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#scraping the data from the page
rank=[]
video_name=[]
artist=[]
upload_date=[]
views=[]
try:
    rk=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rk[:30]:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('-')

try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in name[:30]:
        video_name.append(i.text)
except NoSuchElementException:
    video_name.append('-')

try:
    art=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in art[:30]:
        artist.append(i.text)
except NoSuchElementException:
    artist.append('-')

try:
    upload=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in upload[:30]:
        upload_date.append(i.text)
except NoSuchElementException:
    upload_date.append('-')
    
try:
    view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in view[:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')

#getting the length of each list
print(len(rank))
print(len(video_name))
print(len(upload_date))
print(len(views))
print(len(artist))
```

```python
#creating dataframe
df=pd.DataFrame({'Rank':rank,'Name':video_name,'Artist':artist,'Upload Date':upload_date,'Views in "Mn"':views})

#setting Rank as index
df.set_index('Rank', inplace=True)
df
```
